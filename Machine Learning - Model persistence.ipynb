```python
#import dataset
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
#view features
iris_dataset.feature_names
```

```python
#view target
iris_dataset.target
```

```python
#assign x and y
X_feature = iris_dataset.data
Y_target = iris_dataset.target

```

```python
#create object with new values for prediction 
X_new = [[3,5,4,1],[5,3,4,3]]
```

```python
#using logistic regression estimator
from sklearn.linear_model import LogisticRegression
logReg = LogisticRegression()

```

```python
#fit data into logreg estimator
logReg.fit(X_feature,Y_target)
```

```python
#predict outcome using logreg estimator 
logReg.predict(X_new)
```

```python
#import pickle 
import pickle as pkl
```

```python
#use dumps method to persist the model
persist_model = pkl.dumps(logReg)
persist_model
```

```python
#use joblib to store it as file
import joblib
joblib.dump(logReg,'regresfileone.pkl')
```

```python
#check if the model file is actually saved 
#load the file to a new estimator object 
new_logreg_est = joblib.load('regresfileone.pkl')
new_logreg_est
```

```python
#using new est to predict
new_logreg_est.predict(X_new)
```
